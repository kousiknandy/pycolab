```python
# -*- coding: utf-8 -*-

def update_quality_DEPRECATED(items):
    for item in items:
        if item.name != "Aged Brie" and item.name != "Backstage passes to a TAFKAL80ETC concert":
            if item.quality > 0:
                if item.name != "Sulfuras, Hand of Ragnaros":
                    item.quality = item.quality - 1
        else:
            if item.quality < 50:
                item.quality = item.quality + 1
                if item.name == "Backstage passes to a TAFKAL80ETC concert":
                    if item.sell_in < 11:
                        if item.quality < 50:
                            item.quality = item.quality + 1
                    if item.sell_in < 6:
                        if item.quality < 50:
                            item.quality = item.quality + 1
        if item.name != "Sulfuras, Hand of Ragnaros":
            item.sell_in = item.sell_in - 1
        if item.sell_in < 0:
            if item.name != "Aged Brie":
                if item.name != "Backstage passes to a TAFKAL80ETC concert":
                    if item.quality > 0:
                        if item.name != "Sulfuras, Hand of Ragnaros":
                            item.quality = item.quality - 1
                else:
                    item.quality = item.quality - item.quality
            else:
                if item.quality < 50:
                    item.quality = item.quality + 1
    return items


class Item:
    def __init__(self, name, sell_in, quality):
        self.name = name
        self.sell_in = sell_in
        self.quality = quality

    def __repr__(self):
        return "%s, %s, %s" % (self.name, self.sell_in, self.quality)
```

```python
class OrdinaryUpdater:
    @staticmethod
    def update_item(item):
        if item.quality > 0:
            item.quality -= 1
        item.sell_in -= 1
        if item.sell_in < 0:
            if item.quality > 0:
                item.quality -= 1

class AgedBrieUpdater:
    @staticmethod
    def update_item(item):
        if item.quality < 50:
            item.quality += 1
        item.sell_in -= 1
        if item.sell_in < 0:
            if item.quality < 50:
                item.quality += 1

class SulfurusUpdater:
    @staticmethod
    def update_item(item):
        pass

class BackstageUpdater:
    @staticmethod
    def update_item(item):
        if item.quality < 50:
            item.quality += 1
            if item.sell_in < 11:
                if item.quality < 50:
                    item.quality += 1
            if item.sell_in < 6:
                if item.quality < 50:
                    item.quality += 1
        item.sell_in -= 1
        if item.sell_in < 0:
            item.quality = 0

class UpdaterFactory:
    @staticmethod
    def get_updater(item):
        match item.name:
            case "Aged Brie": return AgedBrieUpdater
            case "Sulfuras, Hand of Ragnaros": return SulfurusUpdater
            case "Backstage passes to a TAFKAL80ETC concert": return BackstageUpdater
            case _: return OrdinaryUpdater

def update_quality(items):
    for item in items:
        updater = UpdaterFactory.get_updater(item)
        updater.update_item(item)
    return items
```
