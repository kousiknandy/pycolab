```python
from random import randrange

gens = []

def thread(thread_id, n):
    k = 0
    print("^", thread_id, n)
    while True:
        if thread_id & (1<<k):
            try:
                r = thread_id & ~(1<<k)
                print(thread_id, ">", n, ">", r)
                gens[r].send(n)
            except StopIteration:
                pass
            print("$", thread_id)
            return
        else:
            m = yield
            n += m
        print("=", thread_id, n)
        k += 1

max_t = 2**5
ns = [randrange(10) for t in range(max_t)]
gens = [thread(t, ns[t]) for t in range(max_t)]
print("=", sum(ns))
for t in range(max_t):
    next(gens[t], None)

```
